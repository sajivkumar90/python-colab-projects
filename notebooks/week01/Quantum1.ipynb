```python
import random

outcomes = ["WIN", "LOSE"]

win_chance = float(input("Enter Chance to win (0.0 to 1.0) : "))

lose_chance = 1.0-win_chance

my_wavefunction = [win_chance, lose_chance]

result = random.choices(outcomes, weights = my_wavefunction, k = 1)

print("Outcome: ",result)
```
